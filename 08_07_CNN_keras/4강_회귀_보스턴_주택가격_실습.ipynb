```python
import pandas as pd

df = pd.read_csv('./house_price.csv')
print(df.shape)
df
```

```python
df = df.dropna(axis=1, how='any')
df
```

```python
df = df.select_dtypes(include=['number'])
df
```

```python
df = df[['YearBuilt', 'YearRemodAdd', 'BsmtFinSF1', 'TotalBsmtSF', '1stFlrSF', 'GrLivArea', 'GarageArea', 'WoodDeckSF', 'SalePrice']]
```

```python
import seaborn as sns
sns.pairplot(df[["YearBuilt", "SalePrice"]])
```

```python
from tensorflow import keras
train_data = df[:1200]
test_data = df[1200:]

train_targets = df['SalePrice'][:1200]
test_targets = df['SalePrice'][1200:]

print(train_data.shape)
```

```python
mean = train_data.mean(axis=0, numeric_only=True)
train_data -= mean
std = train_data.std(axis=0)
train_data /= std
test_data -= mean
test_data /= std
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
train_data = standard_scaler.fit_transform(train_data)
test_data = standard_scaler.transform(test_data)

train_data
```

```python
from keras import models
from keras.layers import Dense
from tensorflow.keras.utils import plot_model
```

```python
boston_model = keras.Sequential([
    Dense(512, input_shape=(9,), activation='relu'),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1)
])
optimizer = keras.optimizers.Adam(learning_rate=0.0001)
boston_model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

# plot_model(boston_model, show_shapes=True, show_layer_activations=True)
```

```python
import numpy as np

history = boston_model.fit(train_data, train_targets, epochs=500, batch_size=32, verbose=1)

test_mse_score, test_mae_score = boston_model.evaluate(test_data, test_targets)

print(test_mae_score)
```

```python
import matplotlib.pyplot as plt

predictions = boston_model.predict(test_data)

plt.scatter(test_targets, predictions)
plt.xlabel('True')
plt.ylabel('Predictions')
plt.grid()
plt.plot([0, 500000], [0, 500000], 'r')
plt.show()
```
