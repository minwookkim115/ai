```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from datasets import Dataset, DatasetDict
```

```python
df = pd.read_csv('/kaggle/input/korean-hate-speech-detection/train.hate.csv')
df.head(10)
```

```python
label2idx = {'none' : 0, 'offensive' : 1, 'hate' : 2}

df['label'] = df['label'].map(lambda x : label2idx[x])
df.head(10)
```

```python
train_dataset = Dataset.from_pandas(df)

train_dataset
```

```python
val_df = pd.read_csv('/kaggle/input/korean-hate-speech-detection/dev.hate.csv')
val_df.head(10)
```

```python
val_df['label'] = val_df['label'].map(lambda x : label2idx[x])
val_df.head(10)
```

```python
val_dataset = Dataset.from_pandas(val_df)
val_dataset
```

```python
from transformers import BertForSequenceClassification, BertTokenizer, Trainer, TrainingArguments
import torch
```

```python
tokenizer = BertTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
def tokenize_function(examples):
    return tokenizer(examples['comments'], padding='max_length', truncation=True)
```

```python
train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)
```

```python
train_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])
val_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])
```

```python
model = BertForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment', num_labels=3, ignore_mismatched_sizes=True)
```

```python
training_args = TrainingArguments(
    output_dir = './results',
    eval_strategy = 'epoch',
    per_device_train_batch_size = 8,
    per_device_eval_batch_size = 8,
    num_train_epochs = 3,
    weight_decay = 0.01,
    logging_dir = './logs',
    logging_steps = 50,
)
```

```python
trainer = Trainer(
    model = model,
    args = training_args,
    train_dataset = train_dataset,
    eval_dataset = val_dataset,
)
```

```python
trainer.train()
```

```python
torch.save(model.state_dict(), "model.pth")
```

```python
model.load_state_dict(torch.load("model.pth"))
```

```python
test_df = pd.read_csv('/kaggle/input/korean-hate-speech-detection/test.hate.no_label.csv')
test_df.head(10)
```

```python
test_dataset = Dataset.from_pandas(test_df)
test_dataset
```

```python
test_dataset = test_dataset.map(tokenize_function, batched=True)
```

```python
predicts = trainer.predict(test_dataset=test_dataset)
```

```python
predict = []

for i in range(len(predicts.predictions)):
    predict.append(predicts.predictions[i].argmax())
```

```python
test_df['label'] = predict
test_df
```

```python
test_df.to_csv('answer.csv', index=False)
```
